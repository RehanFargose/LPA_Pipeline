```python
from datasets import load_dataset
from collections import Counter

# Load in streaming mode to save memory on your Legion Y540
dataset = load_dataset("opennyaiorg/InJudgements_dataset", split="train", streaming=True)

# Fetch the first record to see columns
first_record = next(iter(dataset))
print(f"Dataset Columns: {list(first_record.keys())}")
```

```python
# Dataset Columns: ['Titles', 'Court_Name', 'Cites', 'Cited_by', 'Doc_url', 'Text', 'Doc_size', 'Case_Type', 'Court_Type', 'Court_Name_Normalized']
```

```python
case_types = Counter()
court_types = Counter()

print("Scanning the first 2000 records to identify categories...")

# We iterate through a sample to see what the values look like
for i, record in enumerate(dataset):
    case_types[record['Case_Type']] += 1
    court_types[record['Court_Type']] += 1
    
    if i >= 2000: # Adjust this number to scan more or less
        break

print("\n--- Unique Case Types (Sample) ---")
for k, v in case_types.items():
    print(f"{k}: {v} records")

print("\n--- Unique Court Types (Sample) ---")
for k, v in court_types.items():
    print(f"{k}: {v} records")
```
