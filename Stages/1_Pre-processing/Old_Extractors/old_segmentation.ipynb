```python
import spacy
from spacy.language import Language
import os
```

```python
# Load a model but add custom legal boundaries
nlp = spacy.load("en_core_web_sm")
```

```python
@Language.component("legal_sentence_boundaries")
def set_legal_boundaries(doc):
    for token in doc[:-1]:
        # Prevent splitting on 'Rs.' or 'No.'
        if token.text in ["Rs.", "No.", "Para", "Vol.", "NIA"]:
            doc[token.i + 1].is_sent_start = False
        # Prevent splitting after initials like 'J.' or 'JJ.'
        if token.text in ["J.", "JJ."]:
            doc[token.i + 1].is_sent_start = False
    return doc
```

```python
nlp.add_pipe("legal_sentence_boundaries", before="parser")  
```
