```python
import os
import requests

# 1. Define the original URL and the new 'legal' filename
url = "https://huggingface.co/opennyaiorg/en_legal_ner_trf/resolve/main/en_legal_ner_trf-any-py3-none-any.whl"
valid_filename = "en_legal_ner_trf-3.2.0-py3-none-any.whl"

# 2. Download the file
print("Downloading model...")
response = requests.get(url)
with open(valid_filename, 'wb') as f:
    f.write(response.content)

# 3. Install the renamed local file
print(f"Installing {valid_filename}...")
%pip install {valid_filename}

# 4. Clean up the downloaded file (optional)
# os.remove(valid_filename)
```

```python
import spacy
nlp = spacy.load("en_legal_ner_trf")

# Test with a snippet from your case
text = "The Union encashed bank guarantees to the tune of Rs. 908.91 crores."
doc = nlp(text)

for ent in doc.ents:
    print(f"{ent.text} -> {ent.label_}")
```
