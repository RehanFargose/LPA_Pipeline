```python
import pandas as pd
```

```python
BASE_PATH = f"D:/LPA_MTech_Project/Enriched_Datasets"

# Paths to your parquet files
f1 = f"{BASE_PATH}/SupremeCourt_Combined_1990_2009_enriched.parquet"
f2 = f"{BASE_PATH}/SupremeCourt_Combined_2010_2019_enriched.parquet"
f3 = f"{BASE_PATH}/SupremeCourt_Combined_2020_2025_enriched.parquet"

# D:\LPA_MTech_Project\Enriched_Datasets\SupremeCourt_Combined_1990_2009_enriched.parquet
```

```python
# Load all
df1 = pd.read_parquet(f1)
df2 = pd.read_parquet(f2)
df3 = pd.read_parquet(f3)
```

```python
# Combine
df = pd.concat([df1, df2, df3], ignore_index=True)
```

```python
# Remove duplicates (optional but recommended)
df.drop_duplicates(subset="case_id", keep="first", inplace=True)
```

```python
# Reset index
df.reset_index(drop=True, inplace=True)
```

```python
# Save combined dataset
df.to_parquet(f"{BASE_PATH}/SupremeCourt_Combined_1990_2025_enriched.parquet")
```
