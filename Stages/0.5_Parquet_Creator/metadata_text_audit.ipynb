```python
def check_metadata_by_year(text_folder, metadata_root):
    text_files = [f for f in os.listdir(text_folder) if f.endswith('.txt')]
    
    # Dictionaries to store counts per year
    stats_by_year = {} # Structure: { "year": {"total": 0, "missing": 0} }

    for filename in text_files:
        # Derive year and base_id
        base_id = filename.replace('_EN.txt', '').replace('.txt', '')
        year = base_id.split('_')[0]
        
        # Initialize year in dict if not present
        if year not in stats_by_year:
            stats_by_year[year] = {"total": 0, "missing": 0}
        
        stats_by_year[year]["total"] += 1
        
        # Construct the scattered path
        json_path = os.path.join(metadata_root, year, "metadata", f"{base_id}.json")

        if not os.path.exists(json_path):
            stats_by_year[year]["missing"] += 1

    # Print Summary Table
    print(f"{'Year':<10} | {'Total':<10} | {'Missing':<10} | {'% Missing':<10}")
    print("-" * 50)
    
    # Sort by number of missing files descending
    sorted_years = sorted(stats_by_year.items(), key=lambda x: x[1]['missing'], reverse=True)
    
    for year, counts in sorted_years:
        perc = (counts['missing'] / counts['total']) * 100
        print(f"{year:<10} | {counts['total']:<10} | {counts['missing']:<10} | {perc:>8.2f}%")

    if sorted_years:
        worst_year = sorted_years[0][0]
        print(f"\nYear with the most missing metadata: {worst_year} ({stats_by_year[worst_year]['missing']} files)")
```
