```python
import easyocr
import torch

# Initialize
reader = easyocr.Reader(['en'], gpu=True)

# Check the internal model's device
# We check one of the network modules (like the recognizer)
is_using_gpu = next(reader.recognizer.parameters()).is_cuda

print(f"PyTorch reports CUDA available: {torch.cuda.is_available()}")
print(f"EasyOCR Model is on GPU: {is_using_gpu}")
```
